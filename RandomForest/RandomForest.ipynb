```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Dataset
data = {
    "Age": [18, 22, 25, 30, 35],
    "Buy": [0, 0, 1, 1, 1]
}

# Create DataFrame
df = pd.DataFrame(data)

print(df)

# Features (Input)
X = df[["Age"]]

# Target (Output)
y = df["Buy"]

# Create Model
model = RandomForestClassifier(
    n_estimators=100, # How many trees are created
    random_state=42 # braches of each tree
)

# Train Model
model.fit(X, y)

# New Data for Prediction (Use DataFrame)
new_data = pd.DataFrame({
    "Age": [27]
})

# Prediction
prediction = model.predict(new_data)

# Result
if prediction[0] == 1:
    print("Customer will Buy")
else:
    print("Customer will Not Buy")

# Accuracy
accuracy = model.score(X, y)
print("Accuracy:", accuracy)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# E-commerce Dataset
data = {
    "Age": [20, 25, 30, 35, 40, 28, 32, 45],
    "OrderAmount": [500, 1500, 3000, 4500, 6000, 2000, 3500, 7000],
    "PreviousOrders": [1, 2, 5, 8, 10, 3, 6, 12],
    "Buy": [0, 0, 1, 1, 1, 0, 1, 1]
}

# Create DataFrame
df = pd.DataFrame(data)

print(df)

# Features (Input)
X = df[["Age", "OrderAmount", "PreviousOrders"]]

# Target (Output)
y = df["Buy"]

# Create Model
model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

# Train Model
model.fit(X, y)

# New Customer Data
new_customer = pd.DataFrame({
    "Age": [29],
    "OrderAmount": [2500],
    "PreviousOrders": [9]
})

# Prediction
prediction = model.predict(new_customer)

if prediction[0] == 1:
    print("Customer will Buy")
else:
    print("Customer will Not Buy")

# Accuracy
print("Accuracy:", model.score(X, y))
```
